```python
import cv2
import os
import random
import shutil
```

```python
# Folder containing the video files
video_folder = '../1_Train_Validation_Data/raw/'

path = os.path.abspath('../1_Train_Validation_Data/train/unlabeled_images/')

# Number of frames to extract for each video
num_frames = 1

# Percentage of videos to use for training
train_percentage = 0.7

# Create the training and validation folders
train_folder = os.path.join(path, "train")
print("train_folder", train_folder)
val_folder = os.path.join(path,"val")
print("val_folder", val_folder)
```

```python
# Create a list of all the video files
video_files = os.listdir(video_folder)

# Randomly split the videos into training and validation sets
random.shuffle(video_files)
train_files = video_files[:int(len(video_files)*train_percentage)]
val_files = video_files[int(len(video_files)*train_percentage):]
```

```python
# Iterate over each training video, extract out the frames
for video_file in train_files:
    print("Extracting file from", video_file)
    
    # Read the video
    cap = cv2.VideoCapture(os.path.join(video_folder,video_file))
    
    # Get number of frames in video
    num_frames_total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Randomly select 20 frames
    selected_frames = random.sample(range(num_frames_total),num_frames)
    
    # Iterate over the frames
    for i in selected_frames:
        # Read the frame
        ret, frame = cap.read()
        
        # If frame is read successfully, write it to the training folder
        if ret:
            name = os.path.join(train_folder, video_file +"_frame_" + str(i) + ".jpg")
            cv2.imwrite(name, frame)
    
# Iterate over the validation videos
for video_file in val_files:
    print("Storing validation file from", video_file)
        
    # Save the validation video    
    output_file = os.path.join(val_folder,video_file)
    
    # copy the video file to the validation folder
    shutil.copyfile(os.path.join(video_folder,video_file),output_file)
            
            
```

```python
import cv2
import os

def video_to_frames(video_path, output_folder):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    frame_count = 0

    while True:
        # Read a frame from the video
        ret, frame = video_capture.read()

        # If the frame is not successfully read, we have reached the end of the video
        if not ret:
            break

        # Save the frame as an image
        frame_filename = f"frame_{frame_count:04d}.jpg"  # Frame filenames will be in the format "frame_0000.jpg"
        frame_filepath = os.path.join(output_folder, frame_filename)
        cv2.imwrite(frame_filepath, frame)

        frame_count += 1

    # Release the video capture object
    video_capture.release()

if __name__ == "__main__":
    video_path = "../1_Train_Validation_Data/raw/0021800084-7.mp4"
    output_folder = "../1_Train_Validation_Data/train/unlabeled_images/"

    video_to_frames(video_path, output_folder)
```
