```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load results into Pandas DataFrame
data = pd.read_csv('../3_Training_Validation_Metric_Results/summary results.csv')
```

```python
data.head()
```

```python
# Exclude 'model' and 'epoch' columns
columns_to_plot = data.columns[5:-3]  

# Set Seaborn style
sns.set(style="darkgrid")

# Create a figure and axes for subplots
num_subplots = len(columns_to_plot)

num_cols = 3  # Fixed number of columns
num_rows = (num_subplots + 3) // num_cols  # Calculate number of rows needed for a 4-column layout


# Create a figure and axes for subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Flatten the axes array
axes = axes.flatten()

# Loop through each column and create a lineplot for each
for idx, column in enumerate(columns_to_plot):
    ax = axes[idx]
    sns.lineplot(x='epoch', y=column, hue='iteration', data=data, ax=ax)
    ax.set_title(column)
    ax.set_xlabel('Epoch')
    ax.set_ylabel(column)
    ax.legend(loc='upper right', title='Iterations')  # Add title to the legend

# Adjust layout and spacing
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Remove all white spaces in column names
data = data.rename(columns = lambda x:x.strip())
```

```python
data.columns
```

```python
data.columns[5:9]
```

```python
# Print out the graph for the best metric result in each iteration
for column in data.columns[5:9]:
    iterations = data.groupby('iteration')[column].max()
    plt.figure()
    plt.plot(iterations.index, iterations.values)
    plt.title(f'Plot of best {column} against iterations')
    plt.xlabel('Iterations')
    plt.ylabel(f'Best {column}')
```

```python
# Derive the F1 score base on precision and recall
data['f1_score'] = data['metrics/precision(B)']*data['metrics/recall(B)'] / (data['metrics/precision(B)']+data['metrics/recall(B)'])
data.head()
```

```python
# Print the f1 score
f1_score = data.groupby('iteration')['f1_score'].max()
plt.figure()
plt.plot(f1_score.index, f1_score.values)
plt.title('F1 score vs Iterations')
plt.xlabel('Iterations')
plt.ylabel('F1 Score')
```
